**Loading and Plotting Original Image**

```python
image = cv2.imread('balloon.jpg')  # Load your image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Plot the image
plt.show()
```

**Convert to Grayscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))  # Plot the image
plt.show()
```

**Invert Image:**

```python
inverted_image = 255 - gray_image
plt.imshow(cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB))  # Plot the image
plt.show()
```

**Create Pencil Sketch**

```python
pencil_sketch = cv2.divide(gray_image, inverted_image, scale=256.0)
```

**Display Pencil Sketch:**

```python
plt.imshow(pencil_sketch, cmap='gray')
plt.show()
```
